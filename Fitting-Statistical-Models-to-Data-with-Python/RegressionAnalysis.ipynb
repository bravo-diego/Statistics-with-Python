### Questions 5-6

The next two questions will involve the NHANES dataset.

```python
df = pd.read_csv('nhanes_2015_2016.csv') # read the csv file
df.head() # data preview
```

```python
df.keys() # column names
```

### Preprocessing Data

```python
df['RIAGENDRx'] = df.RIAGENDR.replace({1: 'Male', 2: 'Female'}) 
df['smq'] = df.SMQ020.replace({2: 0, 7: np.nan, 9: np.nan})
df['DMDEDUC2x'] = df.DMDEDUC2.replace({1: 'lt9', 2: 'x9_11', 3: 'HS', 4: 'SomeCollege',
                                       5: 'College', 7: np.nan, 9: np.nan}) # recoding categorical variables
```

```python
log_regression_model = sm.GLM.from_formula('smq ~ RIAGENDRx + RIDAGEYR + DMDEDUC2x', 
                                              family = sm.families.Binomial(), data = df) # fitting a logistic regression model
result = log_regression_model.fit() # estimate model parameters
result.summary()
```

#### Question 5: Which of our predictors has the largest coefficient?

The variable `DMDEDUC2x[T.x9_11]`.

#### Question 6: Which values for DMDEDUC2x and RIAGENDRx are represented in our intercept, or what is our reference level?

**Female** and **College** are the reference levels.

#### Question 7: What model should we use when our target outcome, or dependent variable is binary, or only has two outputs, 0 and 1.

A *logistic regression* model.
